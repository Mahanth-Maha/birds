```python
import pandas as pd
ddg_df = pd.read_csv('assets/duckduckgo_out.csv',names=['Common Names','links','sites_dict','sites_csv'])
ddg_df.head(4)
```

```python
unq_sites = ddg_df.sites_csv
```

```python
from collections import Counter
rem = ('','donttrack.us','twitter.com','reddit.com','vimeo.com','facebook.com')
ddg_list = []
for i in unq_sites:
    p = i.strip().split(',')
    k = [ _ for _ in p if _ not in rem ]
    # print(k)
    ddg_list.append(k)
ddg_list[1:2]
```

```python
total = len(ddg_list)
C = Counter(ddg_list[0])
for i in range(1,total):
    C += Counter(ddg_list[i])
Cd = dict(C)
len(Cd)
```

```python
for i in C.most_common(15):
    perc = (i[1]/total)*100
    print(f'Got {i[1]} out of {total} = {perc}\t\tFor {i[0]}')
```

```python
d = {}
for site in Cd:
    perc = (Cd[site]/total)*100
    print(f'For {site} Got {Cd[site]} out of {total} = {perc}')
```
