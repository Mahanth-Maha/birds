```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from anuvaad import Anuvaad
anu = Anuvaad("english-telugu")
se = "/kaggle/input/birdsdata/Kaggle_birds_out.pkl"
data = pd.read_pickle(se)
# sec = 'data_v3_1_des_anuvaad.'
# data = pd.read_pickle(sec)
def anu_vaad_onebyone(para):
    if len(para) != 0:
        lis_para = [i.strip() for i in para.strip().split('.') if i != '']
        anu_lis_para =  anu.anuvaad(lis_para)
        trans_para = '.'.join(anu_lis_para) + '.'
        return trans_para
    return np.nan

def create_telugu_col_new_optim(col,data):
    di = {}
    data[col+ ' Telugu'] = np.nan
    for i in tqdm(range(data.shape[0])):
        # print(i)
        if type(data[col][i]) == float:
            data[col + ' Telugu'][i] = np.nan
        else:
            if data[col][i] in di:
                data[col + ' Telugu'][i] = di[data[col][i]]
            else:
                data[col + ' Telugu'][i] = anu_vaad_onebyone(data[col][i])
                di[data[col][i]] = data[col + ' Telugu'][i]
    return data
```

```python
data = create_telugu_col_new_optim('Eol_Description',data)
data.to_pickle('data_v3_2_eol_des_anuvaad.pkl')
data.to_csv('data_v3_2_eol_des_anuvaad.csv')
data.to_excel('data_v3_2_eol_des_anuvaad.xlsx')
```
