```python
from googletrans import Translator
def trans_tel(text):
    tr = Translator()
    t = tr.translate(text,dest='telugu',src='english')
    return t.text
trans_tel('just like animals')
```

```python
anu.anuvaad('maha ane nenu...')
```

```python
trans_tel('flight')
```

```python
anu.anuvaad('flight')
```

```python
def comp_trans(sample_text = 'ram ram setji'):
    gpre = trans_tel(sample_text)
    apre = anu.anuvaad(sample_text)
    print('Google : ',gpre)
    print('Anuvaad : ',apre)
    if gpre != apre:
        print("BOTH R NOT SAME 🦁 !") 
    else:
        print('🟰 | TIE ')
    return (gpre,apre)
comp_trans()
```

```python
comp_trans('forest')
```

```python
comp_trans('Pellorneidae')
```

```python
# %pip install tqdm
from tqdm import tqdm
def create_telugu_col(col = 'Binomial Name',data=data):
    data[col+ ' Telugu'] = np.nan
    for i in tqdm(range(data.shape[0])):
        # print(i)
        if type(data[col][i]) == float:
            data[col + ' Telugu'][i] = np.nan
        else:
            data[col + ' Telugu'][i] = anu.anuvaad(data[col][i])
    return data

data = create_telugu_col(data=data)
data.head(5)
```
