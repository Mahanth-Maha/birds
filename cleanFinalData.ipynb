```python
import re
import numpy as np
import string
def clean_cs(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    if type(text) == str and text.lower().strip() not in ('Not evaluated'.lower(),'Data Deficient'.lower(),'',' ','cites appendix ii','data deficient','range restricted'):
        text = text.lower()
        text = re.sub("[\(\[].*?[\)\]]", "", text)
        text = re.sub('\n', '', text)
        # text = re.sub('\[.*?\]', '', text)
        # text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
        text = re.sub('[-‘’“”…]', '', text)
        # text = re.sub(' ', '', text)
        text = re.sub('[0-9]', '', text)
        if 'least concern' in text or 'lc' in text:
            return 'least concern'
        elif 'wild' in text:
            return 'Extinct in the Wild'
        elif 'extinct' in text:
            return 'Extinct'
        elif 'appendix' in text:
            return np.nan
        return text.strip()
    return np.nan
clean_c = lambda x: clean_cs(x)
old.rename({'Conservation status':'cs'},axis=1,inplace=True)
old['Conservation status'] = old.cs.apply(clean_c)
old['Conservation status'] = old.cs.apply(clean_c)
old.drop(['cs'],axis=1,inplace=True)
k =[]
for i in list(old['Conservation status']) :
    if type(i) != float:
        k.append(i)
set(k)
```

```python
old.reset_index(inplace=True)
for i in tqdm(range(old.shape[0])):
    if not pd.isna(old['Synonyms'][i]):
        s_low = old['Synonyms'][i]
        # s_low = s_low.lower()
        bird_low = old['Bird Original Name'][i]
        bird_low = bird_low.lower()
        s_low = s_low.split(',')
        new_s = []
        for j in s_low:
            if j.lower() not in bird_low:
                new_s.append(j)
        old['Synonyms'][i] = ','.join(new_s)
old.set_index('Bird Original Name',inplace=True)
old.sort_index(inplace=True)
old.head(10)
```

```python
from anuvaad import Anuvaad
anu = Anuvaad("english-telugu")

for i in tqdm(range(data.shape[0])):
    if not pd.isna(data['Synonymous Name'][i]):
        s_low = data['Synonymous Name'][i]
        # s_low = s_low.lower()
        bird_low = data['Bird Original Name'][i]
        bird_low = bird_low.lower()
        s_low = s_low.split(',')
        new_s = []
        for j in s_low:
            if j.lower().strip() not in bird_low.strip():
                new_s.append(j)
        lis_syn = ','.join(new_s)
        data['Synonymous Name'][i] = lis_syn
        data['Synonymous Name Telugu'][i] = anu.anuvaad(lis_syn)
data.head(10)
```
