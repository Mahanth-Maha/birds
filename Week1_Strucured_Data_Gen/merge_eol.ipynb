#### Opening All Data sets

```python
birds_csv = pd.read_csv('../assets/birds.csv',index_col='ID')
birds_csv.head(2)
```

```python
birds_dataset_excel = pd.read_excel('../assets/DCPP_Group_Assignment/Birds_Dataset.xlsx',index_col='Serial No')
birds_dataset_excel.head(2)
```

```python
for i in range(1,birds_dataset_excel.shape[0]) :
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("'","")
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("ʻ","")
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("`","")
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("-","")
birds_dataset_excel.tail(3)
```

```python
for i in range(1,birds_csv.shape[0]) :
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("'","")
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("ʻ","")
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("`","")
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("-","")
    

birds_csv.to_csv('../assets/names.csv',columns=['Common Bird Names'])
```
