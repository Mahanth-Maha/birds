### Clean : Wikipedia.org data

```python
wiki_merged = pd.read_csv('../assets/Wiki_Merged.csv',index_col='Common Name')
wiki_merged.drop(['ID'],axis=1,inplace=True)
```

```python
import pickle
import re
import string

def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.lower()
    text = re.sub('\n', '', text)
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('[-‘’“”…]', '', text)
    return text

clean = lambda x: clean_text(x)
```

```python
wiki_merged.reset_index(inplace=True)
```

```python
wiki_merged.rename({"Common Name":"name"},axis=1,inplace=True)
```

```python
wiki_merged.head(3)
```

```python
wiki_merged['name'] = wiki_merged.name.apply(clean)
wiki_merged.head(8)
```

```python
def new_clean_text(text):
    ''' remove Spaces '''
    text = text.lower()
    text = re.sub(' ', '', text)
    return text

clean_space = lambda x: new_clean_text(x)
```

```python
wiki_merged['name_to_merge'] = wiki_merged.name.apply(clean_space)
```

```python
wiki_merged.set_index('name',inplace=True)
```

```python
wiki_merged['name_to_merge'] = wiki_merged.name_to_merge.apply(clean_space)
```
