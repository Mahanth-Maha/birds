```python
def counter_col(wiki):
    wiki_total = wiki.shape[0]
    wiki_d = {}
    for i in wiki.columns:
        wiki_d[i] = wiki_total- wiki[i].isnull().sum()
    wiki_d = sorted(wiki_d.items(), key=lambda x:x[1],reverse=True)
    # return wiki_d
    wiki_df = pd.DataFrame(wiki_d,columns=['Attribute','Occur'])
    wiki_df['Perc'] = (wiki_df['Occur']/wiki_total)*100  
    return wiki_df
# drop list ðŸ˜£
# counter_col(names_eol)
dropping = list(counter_col(names_eol).loc[:,'Attribute'])[21:] 
dropping
```

```python
import numpy as np
def merge_maha(df,A,B,C,defa=False):
    if defa:
        defa = B
    else:
        defa = A
    m1 = ~df[A].isna()
    # df['m1'] = m1
    m2 = ~df[B].isna()
    # df['m2'] = m2
    df[C] = np.select([m1, m2, m1 | m2],[df[A],df[B],df[A]],default=df[defa])
    # df = df.drop([A,B])
    return df
```
