```python
import pandas as pd
```

```python
birds_dataset_excel = pd.read_excel('assets/DCPP_Group_Assignment/Birds_Dataset.xlsx',index_col='Serial No')
# birds_dataset_excel.head(2)
```

```python
birds_dataset_excel['Bird Names'] = birds_dataset_excel['Common Name']
```

```python

for i in range(1,birds_dataset_excel.shape[0]) :
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("'","")
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("ʻ","")
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("`","")
    birds_dataset_excel.loc[i].at['Common Name'] = birds_dataset_excel.loc[i].at['Common Name'].replace("-","")
# birds_dataset_excel.tail(3)
```

```python
birds_csv = pd.read_csv('assets/birds.csv',index_col='ID')
birds_csv['Bird Names'] = birds_csv['Common Bird Names']
for i in range(1,birds_csv.shape[0]) :
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("'","")
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("ʻ","")
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("`","")
    birds_csv.loc[i].at['Common Bird Names'] = birds_csv.loc[i].at['Common Bird Names'].replace("-","")
birds_csv.rename(columns={'Common Bird Names':'Common Name'},inplace=True)
```

```python
birds_csv[2240:2250]
```

```python
new_birds_csv = pd.merge(birds_csv,birds_dataset_excel,how='outer',on=['Common Name'],sort=True)
new_birds_csv.shape
```

```python
new_birds_csv.head(4)
```

```python
dropped = new_birds_csv.dropna(thresh=5)
dropped.shape
```

```python
# unq = dropped.subtract(new_birds_csv,axis=0)
# unq.shape
```

```python

final_df = new_birds_csv[~new_birds_csv.Genus.isin(dropped.Genus)]
final_df.head(5)
```

```python
final_df.drop('Common Name', axis=1, inplace=True)
final_df.rename(columns={'Bird Names_x':'Common Name'},inplace=True)
final_df.reset_index(inplace=True)
final_df.set_index('Common Name')
```

```python
final_df.to_csv('assets/wiki_list_to_comp.csv',columns=['Common Name'],index=False)
```
